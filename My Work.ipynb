```python
import os
import cv2
import numpy as np
import pandas as pd
from scipy.ndimage import gaussian_filter
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import DenseNet201, DenseNet121, ResNet50, MobileNet
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, BatchNormalization, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import random

# Eyelash and Noise Detection
def detect_and_mask_eyelashes_and_noise(image):
    _, eyelash_mask = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY_INV)
    edges = cv2.Canny(image, 50, 150)
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    eyelid_mask = np.zeros_like(image)
    for contour in contours:
        if cv2.boundingRect(contour)[3] > image.shape[0] // 10:
            cv2.drawContours(eyelid_mask, [contour], -1, 255, -1)
    combined_mask = cv2.bitwise_or(eyelash_mask, eyelid_mask)
    masked_image = cv2.bitwise_and(image, image, mask=cv2.bitwise_not(combined_mask))
    return masked_image

# Daugman's Operator
def daugman_operator(image, center_x, center_y, radius_range):
    max_gradient = -1
    best_radius = None
    smoothed_image = gaussian_filter(image, sigma=1.0)
    for radius in radius_range:
        try:
            theta = np.linspace(0, 2 * np.pi, 360)
            circle_x = np.clip(center_x + radius * np.cos(theta), 0, image.shape[1] - 1).astype(int)
            circle_y = np.clip(center_y + radius * np.sin(theta), 0, image.shape[0] - 1).astype(int)
            gradients = np.gradient(smoothed_image[circle_y, circle_x])
            gradient_sum = np.abs(gradients).sum()
            if gradient_sum > max_gradient:
                max_gradient = gradient_sum
                best_radius = radius
        except IndexError:
            continue
    return best_radius, max_gradient

# Normalize Iris Using Rubber Sheet Model
def normalize_iris(image, center, radius, output_size=(64, 512)):
    h, w = output_size
    normalized = np.zeros((h, w), dtype=np.uint8)
    for i in range(h):
        for j in range(w):
            theta = 2 * np.pi * j / w
            r = radius * i / h
            x = int(center[0] + r * np.cos(theta))
            y = int(center[1] + r * np.sin(theta))
            if 0 <= x < image.shape[1] and 0 <= y < image.shape[0]:
                normalized[i, j] = image[y, x]
    return normalized

# Detect Iris and Pupil
def detect_iris_and_pupil(image_path, radius_range=(10, 80)):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        return False, None, None, None, None
    cleaned_image = detect_and_mask_eyelashes_and_noise(image)
    blurred_image = cv2.GaussianBlur(cleaned_image, (9, 9), 2)
    circles = cv2.HoughCircles(
        blurred_image,
        cv2.HOUGH_GRADIENT,
        dp=1,
        minDist=50,
        param1=50,
        param2=30,
        minRadius=10,
        maxRadius=50,
    )
    if circles is None:
        return False, None, None, None, None
    circles = np.uint16(np.around(circles))
    center_x, center_y, _ = circles[0][0]
    pupil_radius, _ = daugman_operator(cleaned_image, center_x, center_y, range(10, 50))
    if pupil_radius is None:
        return False, None, None, None, None
    iris_radius, _ = daugman_operator(cleaned_image, center_x, center_y, range(pupil_radius + 20, pupil_radius + 80))
    if iris_radius is None:
        return False, None, None, None, None
    iris_pupil_ratio = iris_radius / pupil_radius
    return True, iris_radius, pupil_radius, iris_pupil_ratio, (center_x, center_y)

# Generate Age Group Based on Iris-Pupil Ratio
def generate_age_group(iris_pupil_ratio):
    if iris_pupil_ratio > 3.0:
        return "5-20"
    elif 2.0 < iris_pupil_ratio <= 3.0:
        return "21-35"
    elif 1.5 < iris_pupil_ratio <= 2.0:
        return "36-50"
    elif 1.2 < iris_pupil_ratio <= 1.5:
        return "51-70"
    return "Unknown"

# Process Images and Extract Features
def process_images(dataset_path, output_csv_path):
    data = []
    normalized_images = []
    total_images = 0
    successful_images = 0
    image_paths = []
    
    for root, _, files in os.walk(dataset_path):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_paths.append(os.path.join(root, file))

    for img_path in image_paths:
        total_images += 1
        success, iris_radius, pupil_radius, iris_pupil_ratio, center = detect_iris_and_pupil(img_path)
        if not success:
            continue
        age_group = generate_age_group(iris_pupil_ratio)
        if age_group == "Unknown":
            continue
        normalized_iris = normalize_iris(cv2.imread(img_path, cv2.IMREAD_GRAYSCALE), center, iris_radius)
        data.append([img_path, iris_radius, pupil_radius, iris_pupil_ratio, age_group])
        normalized_images.append(normalized_iris)
        successful_images += 1
    if successful_images == 0:
        print("No normalized images were processed. Exiting.")
        return [], []
    df = pd.DataFrame(data, columns=["Image Path", "Iris Radius", "Pupil Radius", "Iris-Pupil Ratio", "Age Group"])
    df.to_csv(output_csv_path, index=False)
    print(f"Total Images: {total_images}, Successful: {successful_images}, Failed: {total_images - successful_images}")
    return np.array(normalized_images), np.array(df["Age Group"])

# Create Model
def create_model(model_type, input_shape=(64, 512, 1)):
    if model_type == "DenseNet201":
        base_model = DenseNet201(weights=None, include_top=False, input_shape=input_shape)
    elif model_type == "DenseNet121":
        base_model = DenseNet121(weights=None, include_top=False, input_shape=input_shape)
    elif model_type == "ResNet50":
        base_model = ResNet50(weights=None, include_top=False, input_shape=input_shape)
    elif model_type == "MobileNet":
        base_model = MobileNet(weights=None, include_top=False, input_shape=input_shape)
    else:
        raise ValueError("Invalid model type. Choose from DenseNet201, DenseNet121, ResNet50, or MobileNet.")
    
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(512, activation="relu")(x)
    x = Dropout(0.2)(x)
    x = Dense(256, activation="relu")(x)
    x = Dropout(0.2)(x)
    predictions = Dense(4, activation="softmax")(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer=Adam(learning_rate=0.001), loss="categorical_crossentropy", metrics=["accuracy"])
    return model

# Train Model
def train_model(X_images, y, model_type, epochs=30, batch_size=32):
    from sklearn.preprocessing import LabelEncoder
    from tensorflow.keras.utils import to_categorical

    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)
    y_categorical = to_categorical(y_encoded)

    X_train, X_val, y_train, y_val = train_test_split(X_images, y_categorical, test_size=0.2, random_state=42)
    model = create_model(model_type, input_shape=X_images.shape[1:])
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        verbose=1
    )
    model.save(f"{model_type}_age_group_prediction.h5")
    print(f"Model saved as {model_type}_age_group_prediction.h5")

    # Plot Training History
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_type} Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid()
    plt.show()

    plt.figure(figsize=(12, 6))
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_type} Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()
    plt.show()

    # Print training and validation accuracy and loss for each epoch
    for i in range(epochs):
        print(f"Epoch {i+1}/{epochs}")
        print(f"Train Accuracy: {history.history['accuracy'][i]:.4f}, Train Loss: {history.history['loss'][i]:.4f}")
        print(f"Validation Accuracy: {history.history['val_accuracy'][i]:.4f}, Validation Loss: {history.history['val_loss'][i]:.4f}\n")

    return model, history

# Main Function
def main():
    # Specify dataset path and output CSV path
    dataset_path = "/kaggle/input/iris-thousand-casia/CASIA-Iris-Thousand"  # Replace with your dataset path
    output_csv_path = "iris_pupil_features_16k.csv"

    if not os.path.exists(dataset_path):
        print(f"Dataset path does not exist: {dataset_path}")
        return

    print(f"Processing images from {dataset_path}...")

    # Step 1: Process images and extract features
    normalized_images, age_groups = process_images(dataset_path, output_csv_path)

    if len(normalized_images) == 0:
        print("No valid images found. Exiting.")
        return

    print(f"Successfully processed {len(normalized_images)} images.")

    # Normalize image data and expand dimensions for compatibility with Keras models
    normalized_images = np.expand_dims(normalized_images.astype("float32") / 255.0, axis=-1)

    # If using MobileNet or ResNet50, convert grayscale to 3 channels
    if normalized_images.shape[-1] == 1:
        normalized_images = np.repeat(normalized_images, 3, axis=-1)  # Convert to 3 channels

    # Step 2: Train multiple models
    models = ["DenseNet201", "DenseNet121", "ResNet50", "MobileNet"]
    for model_type in models:
        print(f"\n--- Training {model_type} ---")
        try:
            train_model(normalized_images, age_groups, model_type=model_type, epochs=30, batch_size=32)
        except Exception as e:
            print(f"Error training {model_type}: {e}")

    print("\n--- All models trained successfully ---")

if __name__ == "__main__":
    main()
```

```python
import os
import cv2
import numpy as np
import pandas as pd
from scipy.ndimage import gaussian_filter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from tensorflow.keras.applications import DenseNet201, DenseNet121, ResNet50, MobileNet
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, BatchNormalization, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import random

# Eyelash and Noise Detection
def detect_and_mask_eyelashes_and_noise(image):
    _, eyelash_mask = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY_INV)
    edges = cv2.Canny(image, 50, 150)
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    eyelid_mask = np.zeros_like(image)
    for contour in contours:
        if cv2.boundingRect(contour)[3] > image.shape[0] // 10:
            cv2.drawContours(eyelid_mask, [contour], -1, 255, -1)
    combined_mask = cv2.bitwise_or(eyelash_mask, eyelid_mask)
    masked_image = cv2.bitwise_and(image, image, mask=cv2.bitwise_not(combined_mask))
    return masked_image

# Daugman's Operator
def daugman_operator(image, center_x, center_y, radius_range):
    max_gradient = -1
    best_radius = None
    smoothed_image = gaussian_filter(image, sigma=1.0)
    for radius in radius_range:
        try:
            theta = np.linspace(0, 2 * np.pi, 360)
            circle_x = np.clip(center_x + radius * np.cos(theta), 0, image.shape[1] - 1).astype(int)
            circle_y = np.clip(center_y + radius * np.sin(theta), 0, image.shape[0] - 1).astype(int)
            gradients = np.gradient(smoothed_image[circle_y, circle_x])
            gradient_sum = np.abs(gradients).sum()
            if gradient_sum > max_gradient:
                max_gradient = gradient_sum
                best_radius = radius
        except IndexError:
            continue
    return best_radius, max_gradient

# Normalize Iris Using Rubber Sheet Model
def normalize_iris(image, center, radius, output_size=(64, 512)):
    h, w = output_size
    normalized = np.zeros((h, w), dtype=np.uint8)
    for i in range(h):
        for j in range(w):
            theta = 2 * np.pi * j / w
            r = radius * i / h
            x = int(center[0] + r * np.cos(theta))
            y = int(center[1] + r * np.sin(theta))
            if 0 <= x < image.shape[1] and 0 <= y < image.shape[0]:
                normalized[i, j] = image[y, x]
    return normalized

# Detect Iris and Pupil
def detect_iris_and_pupil(image_path, radius_range=(10, 80)):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        return False, None, None, None, None
    cleaned_image = detect_and_mask_eyelashes_and_noise(image)
    blurred_image = cv2.GaussianBlur(cleaned_image, (9, 9), 2)
    circles = cv2.HoughCircles(
        blurred_image,
        cv2.HOUGH_GRADIENT,
        dp=1,
        minDist=50,
        param1=50,
        param2=30,
        minRadius=10,
        maxRadius=50,
    )
    if circles is None:
        return False, None, None, None, None
    circles = np.uint16(np.around(circles))
    center_x, center_y, _ = circles[0][0]
    pupil_radius, _ = daugman_operator(cleaned_image, center_x, center_y, range(10, 50))
    if pupil_radius is None:
        return False, None, None, None, None
    iris_radius, _ = daugman_operator(cleaned_image, center_x, center_y, range(pupil_radius + 20, pupil_radius + 80))
    if iris_radius is None:
        return False, None, None, None, None
    iris_pupil_ratio = iris_radius / pupil_radius
    return True, iris_radius, pupil_radius, iris_pupil_ratio, (center_x, center_y)

# Generate Age Group Based on Iris-Pupil Ratio
def generate_age_group(iris_pupil_ratio):
    if iris_pupil_ratio > 3.0:
        return "5-20"
    elif 2.0 < iris_pupil_ratio <= 3.0:
        return "21-35"
    elif 1.5 < iris_pupil_ratio <= 2.0:
        return "36-50"
    elif 1.2 < iris_pupil_ratio <= 1.5:
        return "51-70"
    return "Unknown"

# Process Images and Extract Features
def process_images(dataset_path, output_csv_path):
    data = []
    normalized_images = []
    total_images = 0
    successful_images = 0
    image_paths = []
    
    for root, _, files in os.walk(dataset_path):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_paths.append(os.path.join(root, file))

    for img_path in image_paths:
        total_images += 1
        success, iris_radius, pupil_radius, iris_pupil_ratio, center = detect_iris_and_pupil(img_path)
        if not success:
            continue
        age_group = generate_age_group(iris_pupil_ratio)
        if age_group == "Unknown":
            continue
        normalized_iris = normalize_iris(cv2.imread(img_path, cv2.IMREAD_GRAYSCALE), center, iris_radius)
        data.append([img_path, iris_radius, pupil_radius, iris_pupil_ratio, age_group])
        normalized_images.append(normalized_iris)
        successful_images += 1
    if successful_images == 0:
        print("No normalized images were processed. Exiting.")
        return [], []
    df = pd.DataFrame(data, columns=["Image Path", "Iris Radius", "Pupil Radius", "Iris-Pupil Ratio", "Age Group"])
    df.to_csv(output_csv_path, index=False)
    print(f"Total Images: {total_images}, Successful: {successful_images}, Failed: {total_images - successful_images}")
    return np.array(normalized_images), np.array(df["Age Group"])

# Create Model
def create_model(model_type, input_shape=(64, 512, 1)):
    if model_type == "DenseNet201":
        base_model = DenseNet201(weights=None, include_top=False, input_shape=input_shape)
    elif model_type == "DenseNet121":
        base_model = DenseNet121(weights=None, include_top=False, input_shape=input_shape)
    elif model_type == "ResNet50":
        base_model = ResNet50(weights=None, include_top=False, input_shape=input_shape)
    elif model_type == "MobileNet":
        base_model = MobileNet(weights=None, include_top=False, input_shape=input_shape)
    else:
        raise ValueError("Invalid model type. Choose from DenseNet201, DenseNet121, ResNet50, or MobileNet.")
    
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(512, activation="relu")(x)
    x = Dropout(0.2)(x)
    x = Dense(256, activation="relu")(x)
    x = Dropout(0.2)(x)
    predictions = Dense(4, activation="softmax")(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer=Adam(learning_rate=0.001), loss="categorical_crossentropy", metrics=["accuracy"])
    return model

# Train Model
def train_model(X_train, y_train, X_val, y_val, model_type, epochs=30, batch_size=32):
    # Create model based on model_type
    model = create_model(model_type, input_shape=X_train.shape[1:])

    # Train the model
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        verbose=1
    )

    # Save the model
    model.save(f"{model_type}_age_group_prediction.h5")
    print(f"Model saved as {model_type}_age_group_prediction.h5")

    # Plot Training History
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_type} Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_type} Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Evaluate model on validation data to get final test accuracy
    val_loss, val_accuracy = model.evaluate(X_val, y_val, verbose=0)
    print(f"Final Test Accuracy: {val_accuracy:.4f}")

    # Return the final training and validation accuracies
    final_train_accuracy = history.history['accuracy'][-1]
    final_val_accuracy = history.history['val_accuracy'][-1]

    return model, final_train_accuracy, final_val_accuracy

# Plot Confusion Matrix
def plot_confusion_matrix(model, X, y, label_encoder, dataset_type="Test"):
    # Predict the classes on the dataset
    y_pred = model.predict(X)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true = np.argmax(y, axis=1)

    # Generate the confusion matrix
    cm = confusion_matrix(y_true, y_pred_classes)

    # Plot the confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'{dataset_type} Confusion Matrix')
    plt.show()

# Main Function
def main():
    # Specify dataset path and output CSV path
    dataset_path = "/kaggle/input/iris-thousand-casia/CASIA-Iris-Thousand"  # Replace with your dataset path
    output_csv_path = "iris_pupil_features_16k.csv"

    if not os.path.exists(dataset_path):
        print(f"Dataset path does not exist: {dataset_path}")
        return

    print(f"Processing images from {dataset_path}...")

    # Step 1: Process images and extract features
    normalized_images, age_groups = process_images(dataset_path, output_csv_path)

    if len(normalized_images) == 0:
        print("No valid images found. Exiting.")
        return

    print(f"Successfully processed {len(normalized_images)} images.")

    # Normalize image data and expand dimensions for compatibility with Keras models
    normalized_images = np.expand_dims(normalized_images.astype("float32") / 255.0, axis=-1)

    # Convert grayscale to 3 channels if necessary
    if normalized_images.shape[-1] == 1:
        normalized_images = np.repeat(normalized_images, 3, axis=-1)  # Convert to 3 channels

    # Encode the labels
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(age_groups)
    y_categorical = to_categorical(y_encoded)

    # Split the dataset into train and validation (test) sets
    X_train, X_val, y_train, y_val = train_test_split(normalized_images, y_categorical, test_size=0.2, random_state=42)

    # Step 2: Train the model and plot the results
    models = ["DenseNet201", "DenseNet121", "ResNet50", "MobileNet"]

    for model_type in models:
        print(f"\n--- Training {model_type} ---")
        try:
            model, train_acc, val_acc = train_model(X_train, y_train, X_val, y_val, model_type=model_type, epochs=30, batch_size=32)

            # Display the final train and validation accuracies
            print(f"Final Train Accuracy: {train_acc:.4f}")
            print(f"Final Validation Accuracy: {val_acc:.4f}")

            # Plot confusion matrices for train and validation sets
            plot_confusion_matrix(model, X_train, y_train, label_encoder, dataset_type="Train")
            plot_confusion_matrix(model, X_val, y_val, label_encoder, dataset_type="Test")

        except Exception as e:
            print(f"Error training {model_type}: {e}")

    print("\n--- All models trained successfully ---")

if __name__ == "__main__":
    main()
```
